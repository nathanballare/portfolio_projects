## Encontrando a Localização Ideal para um Supermercado no Brasil

É essencial continuar monitorando e avaliando o desempenho das operações em São Paulo enquanto planejamos a expansão. Isso garantirá que a empresa esteja preparada para enfrentar os desafios específicos de diferentes mercados regionais.

Resumindo, este projeto utilizará técnicas de ciência de dados, análise geoespacial e aprendizado de máquina para solucionar o problema de encontrar o melhor local para a abertura de uma rede de supermercados. Função para obter locais próximos (venues) em um bairro específico. Essa função recebe como entrada os nomes dos bairros, suas coordenadas de latitude e longitude, bem como os parâmetros de raio e limite para controlar a pesquisa. Ela cria uma solicitação à API do Google Cloud para obter os locais próximos a cada bairro e processa os resultados para criar um DataFrame com informações relevantes.

```python
def getNearbyVenues(names, latitudes, longitudes, radius=500, venue_type=None):
    
    venues_list = []
    
    for name, lat, lng in zip(names, latitudes, longitudes):
        # Use a biblioteca googlemaps para pesquisar locais próximos com base no tipo especificado
        places = gmaps.places_nearby(location=(lat, lng), radius=radius, type=venue_type)
        
        for place in places['results']:
            venue_name = place['name']
            venue_lat = place['geometry']['location']['lat']
            venue_lng = place['geometry']['location']['lng']
            venue_category = place['types']  # Use as categorias fornecidas pelo Google
            venues_list.append([
                name,
                lat,
                lng,
                venue_name,
                venue_lat,
                venue_lng,
                venue_category
            ])
    
    # Crie o DataFrame nearby_venues diretamente aqui com as colunas corretas
    nearby_venues = pd.DataFrame(venues_list, columns=['Neighbourhood', 
                                                       'Neighbourhood Latitude', 
                                                       'Neighbourhood Longitude', 
                                                       'Venue', 
                                                       'Venue Latitude', 
                                                       'Venue Longitude', 
                                                       'Venue Category'])
    
    return nearby_venues


```

```python
# Tendo os dados dos bairros, latitudes e longitudes em df_sp
# Vamos pesquisar locais dentro de um raio de 1 km do ponto central do bairro
radius = 1000

# Chamar a função getNearbyVenues
saopaulo_venues = getNearbyVenues(names=df_sp['Neighbourhood'],
                                              latitudes=df_sp['Latitude'],
                                              longitudes=df_sp['Longitude'],
                                              radius=radius)

print(saopaulo_venues.shape)
print(saopaulo_venues.head())

```

```python
# Create a new DataFrame with Neighbourhood and Venue Category
saopaulo_onehot = saopaulo_venues[['Neighbourhood', 'Venue Category']]

# Explode the lists in 'Venue Category' into separate rows
saopaulo_onehot = saopaulo_onehot.explode('Venue Category')

# Add a 'Count' column to represent the count of each category
saopaulo_onehot['Count'] = 1

# Pivot the DataFrame to have venue categories as columns
saopaulo_onehot_pivot = pd.pivot_table(saopaulo_onehot, values='Count', index='Neighbourhood', columns='Venue Category', aggfunc='sum', fill_value=0)

# Reset the index to make 'Neighbourhood' a regular column
saopaulo_onehot_pivot.reset_index(inplace=True)

# Display the resulting DataFrame
print(saopaulo_onehot_pivot)
```

Cada linha representa um bairro ('Neighbourhood'), um valor de 1 indica a presença de um determinado comercio nesse bairro, enquanto 0 indica a ausência. Isso é típico de uma representação one-hot encoding, onde cada categoria se torna uma coluna binária separada.


Agrupando linhas do DataFrame saopaulo_onehot_pivot pelo bairro (Neighbourhood) e calculando a média das ocorrências de cada categoria, o que resulta na frequência relativa de cada categoria em cada bairro. A saída exibida deve ter um tamanho igual ao número de bairros e o número de categorias.


```python
saopaulo_grouped = saopaulo_onehot_pivot.groupby('Neighbourhood').mean().reset_index()
print(f'Size: {saopaulo_grouped.shape}')
saopaulo_grouped.head()
```

Segmentação de bairro (Neighbourhood)
Convertendo em um Dataframe

```python
# Defina o número de principais categorias de locais que você deseja para cada bairro
num_top_venues = 10

def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
# Crie uma lista de dicionários para armazenar as categorias mais comuns de estabelecimentos comerciais locais para cada bairro
data = []

# Preencha a lista de dicionários com os dados corretos
for ind in np.arange(saopaulo_grouped.shape[0]):
    neighbourhood = saopaulo_grouped['Neighbourhood'].iloc[ind]
    top_venues = return_most_common_venues(saopaulo_grouped.iloc[ind, :], num_top_venues)
    
    # Crie um dicionário com o nome do bairro e as categorias de estabelecimentos comerciais locais mais comuns
    neighbourhood_data = {'Neighbourhood': neighbourhood}
    for i, category in enumerate(top_venues):
        neighbourhood_data[f'{i+1} Most Common Venue'] = category
    
    # Adicione o dicionário à lista de dados
    data.append(neighbourhood_data)

# Crie um novo DataFrame a partir da lista de dicionários
neighbourhoods_venues_sorted = pd.DataFrame(data)

# Verifique o novo DataFrame
neighbourhoods_venues_sorted.head()

```

```python
import os
os.environ['OMP_NUM_THREADS'] = '1'

# Defina o número de clusters (k)
k = 4

# Selecione apenas as colunas relacionadas às categorias mais comuns de estabelecimentos comerciais locais
saopaulo_grouped_clustering = saopaulo_grouped.drop('Neighbourhood', axis=1)

# Crie uma instância do modelo KMeans
kmeans = KMeans(n_clusters=k, random_state=1, n_init=20)

# Ajuste o modelo aos seus dados
kmeans.fit(saopaulo_grouped_clustering)

# Adicione as etiquetas dos clusters ao DataFrame neighbourhoods_venues_sorted
neighbourhoods_venues_sorted['Cluster Labels'] = kmeans.labels_

# Verifique o DataFrame neighbourhoods_venues_sorted com as etiquetas dos clusters
print(neighbourhoods_venues_sorted.head())
```

Análise Geoespacial:

Agora vamos enocntrar os dados de latitude e longitude das capitais, para criar um mapa interativo que destaque as áreas mais promissoras para a expansão de supermercados. Isso pode ser feito com bibliotecas como Folium. 

```python
# Lista com os nomes das capitais
capitais = [
    'PORTO VELHO', 'RIO BRANCO', 'MANAUS', 'BOA VISTA', 'BELÉM', 'MACAPÁ',
    'PALMAS', 'SÃO LUÍS', 'TERESINA', 'FORTALEZA', 'NATAL', 'JOÃO PESSOA',
    'RECIFE', 'MACEIÓ', 'ARACAJU', 'SALVADOR', 'Belo Horizonte', 'VITÓRIA',
    'RIO DE JANEIRO', 'CURITIBA', 'FLORIANÓPOLIS', 'PORTO ALEGRE',
    'CAMPO GRANDE', 'CUIABÁ', 'GOIÂNIA', 'BRASÍLIA'
]

# Função para obter coordenadas usando geopy
def get_coordinates(city_name):
    geolocator = Nominatim(user_agent="geoapiExercises")
    location = geolocator.geocode(city_name)
    if location:
        return location.latitude, location.longitude
    else:
        return None, None

# Obter coordenadas para as capitais
coordinates = [get_coordinates(city) for city in capitais]
latitudes, longitudes = zip(*coordinates)

# Criar DataFrame com as capitais e suas coordenadas
df_capitais = pd.DataFrame({
    'NOMEMUN': capitais,
    'Latitude': latitudes,
    'Longitude': longitudes
})

# Exibir as primeiras linhas do DataFrame para verificar as coordenadas
print(df_capitais.head())
```

Atribuindo clusters ao DataFrame df_capitais. Certificando se os clusters reais estão atribuídos a cada capital com base na análise de clusters que realizei anteriormente e substituindo o array clusters no código pelo array de clusters reais.
